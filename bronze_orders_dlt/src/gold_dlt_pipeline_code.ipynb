```python
import dlt
from pyspark.sql.functions import current_timestamp, col, count, sum, avg

@dlt.table(
    name="bronze.default.gold_category_metrics", #USED Same Catalog due to ADB UI Issue, can use gold caatalog in ADB CLI
    comment="Gold aggregated metrics by category"
)
def gold_category_metrics():
    return (
        dlt.read("bronze.default.silver_orders")
        .groupBy("category")
        .agg( #aggregation
            count("*").alias("total_orders"),
            sum("amount").alias("total_amount"),
            avg("amount").alias("avg_amount")
        )
        .withColumn("gold_ingestion_date", current_timestamp())
    )
```
